```python
# Initial code
import random
import math
import matplotlib.pyplot as plt
import operator

# Set the pseudo-random seed for reproducibility
random.seed(0)

# A variable to store the number of agents
n_agents = 10

# Initialise agents
agents = []
for i in range(n_agents):
    agents.append([random.randint(0, 99), random.randint(0, 99)])
print(agents)

# Move agents
for i in range(n_agents):
    # Change agents[i] coordinates randomly
    # x-coordinate
    rn = random.random()
    #print("rn", rn)
    if rn < 0.5:
        agents[i][0] = agents[i][0] + 1
    else:
        agents[i][0] = agents[i][0] - 1
    # y-coordinate
    rn = random.random()
    #print("rn", rn)
    if rn < 0.5:
        agents[i][1] = agents[i][1] + 1
    else:
        agents[i][1] = agents[i][1] - 1
print(agents)

# Calculate the Euclidean distance between (x0, y0) and (x1, y1)
# Set x0 and y0 to equal 0, x1 to equal 3, and y1 to equal 4
x0 = 0
y0 = 0
x1 = 3
y1 = 4
# Calculate the difference in the x coordinates.
dx = x0 - x1
# Calculate the difference in the y coordinates.
dy = y0 - y1
# Square the differences and add the squares
ssd = (dx * dx) + (dy * dy)
print("ssd", ssd)
# Calculate the square root
distance = ssd ** 0.5
print("distance", distance)
distance = math.sqrt(ssd)
print("distance", distance)

# Plot
for i in range(n_agents):
    plt.scatter(agents[i][0], agents[i][1], color='black')
# Plot the coordinate with the largest x red
lx = max(agents, key=operator.itemgetter(0))
plt.scatter(lx[0], lx[1], color='red')
# Plot the coordinate with the smallest x blue
sx = min(agents, key=operator.itemgetter(0))
plt.scatter(sx[0], sx[1], color='blue')
# Plot the coordinate with the largest y yellow
ly = max(agents, key=operator.itemgetter(1))
plt.scatter(ly[0], ly[1], color='yellow')
# Plot the coordinate with the smallest y green
sy = min(agents, key=operator.itemgetter(1))
plt.scatter(sy[0], sy[1], color='green')
plt.show()
```

```python
def get_distance(x0, y0, x1, y1):
    distance = math.sqrt((x1 - x0)**2 + (y1 - y0)**2)
    return distance

def get_max_distance(agents):
    max_distance = 0
    for i in range(len(agents)):
        for j in range(i + 1, len(agents)):
            distance = math.sqrt((agents[j][0] - agents[i][0])**2 + (agents[j][1]-agents[i][1])**2)
            max_distance = max(distance, max_distance)
    
    return max_distance

def get_min_distance(agents):
    min_distance = math.inf
    for i in range(len(agents)):
        for j in range(i + 1, len(agents)):
            distance = math.sqrt((agents[j][0] - agents[i][0])**2 + (agents[j][1]-agents[i][1])**2)
            min_distance = min(distance, min_distance)
    
    return min_distance

def get_min_max_distance(agents):
    min_distance = math.inf
    max_distance = 0
    for i in range(len(agents)):
        for j in range(i + 1, len(agents)):
            distance = math.sqrt((agents[j][0] - agents[i][0])**2 + (agents[j][1]-agents[i][1])**2)
            min_distance = min(distance, min_distance)
            max_distance = max(distance, max_distance)
    
    return (max_distance, min_distance)

import time

start = time.perf_counter()
min_d = get_min_distance(agents)
max_d = get_max_distance(agents)
end = time.perf_counter()

print("Time taken to calculate maximum distance", end - start, "seconds")

start = time.perf_counter()
min_max_d = get_min_max_distance(agents)
end = time.perf_counter()

print("Time taken to calculate maximum distance", end - start, "seconds")
```

```python
import numpy as np
import statistics

def get_statistic_variable(agents):
    distances = []
    for i in range(len(agents)):
        for j in range(i + 1, len(agents)):
            distance = math.sqrt((agents[j][0] - agents[i][0])**2 + (agents[j][1]-agents[i][1])**2)
            distances.append(distance)
    
    distance.sort()
    distance_mean = np.mean(distances)
    distance_std = np.std(distances)
    distance_median = np.median(distances)
    distance_mode = statistics.mode(distance)

    # Calculate truncated mean
    p = 0.05
    lower, upper = np.percentile(distances, [p * 100, (1-p) * 100])
    distance_truncated_mean = np.mean(distances[(distances >= lower) & (distances <= upper)])

    final_result = {'mean':distance_mean, 
                    'std':distance_std, 
                    'median':distance_median, 
                    'mode':distance_mode, 
                    'truncated mean':distance_truncated_mean}

    return final_result
```

```python
n_agents = 10
new_agents = []
for i in range(n_agent):
        new_agents.append([random.randint(0, 99) for i in range(2)])
n_iterations = 1000
# Variables for constraining movement.
# The minimum x coordinate.
x_min = 0
# The minimum y coordinate.
y_min = 0
# The maximum x coordinate.
x_max = 99
# The maximum y coordinate.
y_max = 99

for agent in new_agents:
        for i in range(n_iterations):
                # Apply movement constraints.
                if agent[0] < x_min:
                    agent[0] = x_min
                if agent[1] < y_min:
                    agent[1] = y_min
                if agent[0] > x_max:
                    agent[0] = x_max
                if agent[1] > y_max:
                    agent[1] = y_max
```
