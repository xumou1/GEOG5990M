```python
import csv
import my_modules.agentframework as af
import my_modules.io as io
import matplotlib.pyplot as plt
import numpy as np
import operator

environment = io.read_data()
n_rows, n_cols = np.shape(environment)
plt.imshow(environment)
```

```python
agents = []
n_agents = 10
for i in range(n_agents):
    # Create an agent
    agents.append(af.Agent(i, environment, n_rows, n_cols))
    print(agents[i])

for agent in agents:
    plt.scatter(agent.x, agent.y, color = "black")

# Plot the coordinate with the largest x red
lx = max(agents, key=operator.attrgetter('x'))
plt.scatter(lx.x, lx.y, color='red')
# Plot the coordinate with the smallest x blue
sx = min(agents, key=operator.attrgetter('x'))
plt.scatter(sx.x, sx.y, color='blue')
# Plot the coordinate with the largest y yellow
ly = max(agents, key=operator.attrgetter('y'))
plt.scatter(ly.x, ly.y, color='yellow')
# Plot the coordinate with the smallest y green
sy = min(agents, key=operator.attrgetter('y'))
plt.scatter(sy.x, sy.y, color='green')
# The maximum an agents x coordinate is allowed to be.
x_min = 0
x_max = n_cols - 1
# The maximum an agents y coordinate is allowed to be.
#y_max = 99
y_min = 0
y_max = n_rows - 1
n_iterations = 10
for agent in agents:
    for i in range(n_iterations):
        # Apply movement constraints.
        agent.eat()
        agent.move(x_min, y_min, x_max, y_max)
    plt.scatter(agent.x, agent.y, color = "red")
plt.ylim(y_max / 3, y_max * 2 / 3)
plt.xlim(x_max / 3, x_max * 2 / 3)
plt.imshow(environment)

print(sum_enviroment(environment))
print(sum_agent_store(agents))
```

```python
def sum_enviroment(environment):
    return np.sum(environment)

def sum_agent_store(agents):
    sum_store = 0
    for agent in agents:
        sum_store += agent.store
    return sum_store
```
